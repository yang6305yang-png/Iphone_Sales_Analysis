# Top 10 highest-rated iPhones on Flipkart in India?

```python
data.head()
```

```python
highest_rated = data.sort_values(by= ["Star Rating"], ascending= False)

highest_rated = highest_rated.head(10)
```

```python
highest_rated.head(10)
```

# How many ratings do the highest-rated iPhones on Flipkart have?

```python
iphones = highest_rated['Product Name'].value_counts()
label = iphones.index
counts = highest_rated["Number Of Ratings"]
figure = px.bar(highest_rated, x = label, y = counts)
figure.show()
```

# Which iPhone has the highest number of revies on Flipkart

```python
data.head()
```

```python
iphones = highest_rated['Product Name'].value_counts()
label = iphones.index
counts = highest_rated['Number Of Reviews']
figure = px.bar(highest_rated, x = label, y = counts, title= 'number of reviews of Higest Rated iPhones')
figure.show()
```

# What is the relationship between the sale price of iPhone and the number

```python
data.head()
```

```python
import plotly.express as px

figure = px.scatter(
    data_frame=data,
    x="Number Of Ratings",
    y="Sale Price",
    size="Discount Percentage",
    trendline="ols",
    title="Relationship between Sales and Number of Ratings"
)

figure.show()
```

 There is a negative linear relationship between the sale price of iPhones and the number of ratings. It means iPhone with lower sale prices are sold more in India 

# What is the relationship between the discount percentage and the number of ratings on Flipkart?

```python
figure = px.scatter(data_frame= data, x = "Number Of Ratings",
                    y = "Discount Percentage", size = "Sale Price",
                    trendline = 'ols',
                    title = "Relationship between | and Number of Ratings of iPhones")
figure.show()
```

# Most expensive and least expensive

```python
most_expensive = data.loc[data['Sale Price'].idxmax()]
least_expensive = data.loc[data['Sale Price'].idxmin()]

# Display the results
print("Most Expensive Product:")
print(most_expensive)

print("\nLeast Expensive Product:")
print(least_expensive)
```
